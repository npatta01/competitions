```python
import cv2
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
import matplotlib.cm as cm

import helper

import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = (20, 20)  # that's default image size for this 
from skimage.feature import hog
from skimage import data, color, exposure

%matplotlib inline
```

```python
sample_whale_img='../../data/imgs/w_10005.jpg'
```

```python
im = plt.imread(sample_whale_img)
```

```python
plt.imshow(im)
```

```python
only_whale_img=im[900:1500,900:2000,:]
plt.imshow(only_whale_img)
plt.axis('off')
```

```python
only_whale_img_gray = color.rgb2gray(only_whale_img)
plt.imshow(only_whale_img_gray,cmap = cm.Greys_r)
plt.axis('off')
```

```python
fd, hog_image = hog(only_whale_img_gray, orientations=8, pixels_per_cell=(16, 16),
                    cells_per_block=(1, 1), visualise=True)
```

```python
hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 0.02))
```

```python
plt.imshow(hog_image_rescaled,cmap = cm.Greys_r)
plt.axis('off')
```
